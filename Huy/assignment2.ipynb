```python
import matplotlib.pyplot as plt
import numpy as np

# Tạo dữ liệu ngẫu nhiên
np.random.seed(10)
data = np.random.normal(loc=0, scale=1, size=100)

# Vẽ biểu đồ box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.title('Biểu đồ Box Plot')
plt.ylabel('Giá trị')
plt.show()
```
